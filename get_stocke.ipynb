```python
from nsetools import Nse
import pandas as pd
```

```python
nse = Nse()
```

```python
get_stocks_in_index = nse.get_stocks_in_index()
df = pd.DataFrame(get_stocks_in_index)
df
```

```python
top_gainers = nse.get_top_gainers()  # may return dict or list of stock codes
df = pd.DataFrame(top_gainers)
df
```

```python
week_52_high = nse.get_52_week_high()
df = pd.DataFrame(week_52_high)
df
```

```python
week_52_low = nse.get_52_week_low()
df = pd.DataFrame(week_52_low)
df
```

```python
stock_codes = nse.get_stock_codes()
df = pd.DataFrame(stock_codes)
df.head()
```

```python
get_quote = nse.get_quote("AIMTRON")
df = pd.DataFrame([get_quote])
df
```

```python
get_future_quote = nse.get_future_quote("RELIANCE")
df = pd.DataFrame(get_future_quote)
df
```
