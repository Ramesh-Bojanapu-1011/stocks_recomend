```python
import requests
from io import BytesIO, StringIO

# Some servers (like BSE) block simple programmatic requests; send a browser-like User-Agent
try:
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0 Safari/537.36"
    }
    resp = requests.get("https://www.bseindia.com/downloads1/List_of_companies.csv", headers=headers, timeout=15)
    resp.raise_for_status()
    content = resp.content
    # Try decode as UTF-8 text first, otherwise read as bytes
    try:
        text = content.decode("utf-8")
        df = pd.read_csv(StringIO(text))
    except Exception:
        df = pd.read_csv(BytesIO(content))


except Exception as e:
    print("Failed to download or parse CSV from", "https://www.bseindia.com/downloads1/List_of_companies.csv", "->", repr(e))
    df = pd.DataFrame()
df
```
